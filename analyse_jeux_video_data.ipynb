```python
import pandas as pd

# Lecture du fichier CSV (place-le dans le même dossier que le notebook)
df = pd.read_csv("jeux_video_dataset.csv")

# Affiche les premières lignes
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Taille des graphiques
plt.figure(figsize=(15, 5))

# Histogramme des scores critiques
plt.subplot(1, 3, 1)
sns.histplot(df['Critic_Score'], kde=True, bins=5, color="skyblue")
plt.title("Répartition des notes critiques")
plt.xlabel("Critic Score")

# Histogramme des scores utilisateurs
plt.subplot(1, 3, 2)
sns.histplot(df['User_Score'], kde=True, bins=5, color="lightgreen")
plt.title("Répartition des notes utilisateurs")
plt.xlabel("User Score")

# Histogramme des ventes
plt.subplot(1, 3, 3)
sns.histplot(df['Global_Sales_M'], kde=True, bins=5, color="salmon")
plt.title("Répartition des ventes mondiales (en millions)")
plt.xlabel("Ventes mondiales")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Year")
plt.title("Nombre de jeux par année")
plt.xlabel("Année")
plt.ylabel("Nombre de jeux")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# On ne garde que les colonnes numériques
corr = df.corr(numeric_only=True)

# Affichage de la heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```

```python
# Visualisation du nombre de jeux vidéo par genre
# Cela permet d’identifier quels genres sont les plus répandus dans le dataset (Action, Sports, etc.)
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Genre", order=df["Genre"].value_counts().index, palette="Set2")
plt.xticks(rotation=45)
plt.title("Répartition des jeux par genre")
plt.xlabel("Genre")
plt.ylabel("Nombre de jeux")
plt.tight_layout()
plt.show()
```

```python
# Analyse de la répartition des jeux par plateforme (PS4, Switch, PC, etc.)
# Utile pour voir quelles plateformes accueillent le plus de jeux ou sont les plus populaires
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x="Platform", order=df["Platform"].value_counts().index, palette="Set3")
plt.xticks(rotation=45)
plt.title("Répartition des jeux par plateforme")
plt.xlabel("Plateforme")
plt.ylabel("Nombre de jeux")
plt.tight_layout()
plt.show()
```

```python
# Affichage des 10 jeux ayant généré le plus de ventes à l’échelle mondiale
# Permet de repérer les titres les plus influents commercialement
top10 = df.sort_values(by="Global_Sales_M", ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(data=top10, y="Name", x="Global_Sales_M", palette="viridis")
plt.title("Top 10 des jeux les plus vendus")
plt.xlabel("Ventes globales (millions)")
plt.ylabel("Nom du jeu")
plt.tight_layout()
plt.show()
```

```python
# Distribution des notes données par les critiques (Critic_Score)
# Donne une idée de la perception qualitative des jeux dans le dataset
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x="Critic_Score", kde=True, bins=10, color="skyblue")
plt.title("Distribution des notes critiques")
plt.xlabel("Critic Score")
plt.ylabel("Nombre de jeux")
plt.show()
```

```python
# Somme des ventes mondiales de jeux vidéo par année de sortie
# Permet de visualiser l’évolution du marché dans le temps
plt.figure(figsize=(10, 5))
sns.barplot(data=df, x="Year", y="Global_Sales_M", estimator=sum, palette="coolwarm")
plt.title("Ventes globales par année")
plt.xlabel("Année")
plt.ylabel("Ventes (en millions)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## 🔍 Résumé de l'analyse exploratoire (EDA)

### 🎯 Objectif
L’objectif de cette analyse était de comprendre les tendances du marché des jeux vidéo en étudiant des variables comme l’année, le genre, la plateforme, les scores critiques et les ventes.

---

### 📅 Nombre de jeux par année
- Le nombre de jeux publiés varie fortement selon les années.
- On remarque une **baisse des sorties dans les années les plus récentes**, probablement parce que les données sont incomplètes ou que les jeux récents sont moins nombreux.

---

### 🎮 Genres de jeux les plus représentés
- Les genres les plus courants sont **Action**, **Sports** et **Shooter**.
- Des genres comme **Puzzle**, **Fighting** ou **Strategy** sont **moins représentés**, suggérant une popularité ou un investissement moindre.

---

### 💻 Répartition par plateforme
- Certaines plateformes dominent très nettement en volume de jeux (ex : **PS2**, **DS**, **PS4** selon les données).
- Les plateformes récentes peuvent avoir moins de jeux simplement parce que le dataset est centré sur une période donnée.

---

### 🏆 Top 10 des jeux les plus vendus
- Les jeux les plus vendus sont souvent issus de **franchises bien établies** (ex : *Wii Sports*, *Mario Kart*, *GTA*).
- On retrouve principalement des jeux sortis sur des **plateformes populaires** et avec une **forte visibilité marketing**.

---

### ⭐ Distribution des notes critiques
- Les notes critiques sont **globalement concentrées entre 60 et 90**, ce qui indique une tendance à l’évaluation **positive** des jeux dans ce dataset.
- Peu de jeux reçoivent des notes très faibles ou parfaites, ce qui est attendu.

---

### 💰 Ventes globales par année
- Les années autour de **2008-2012** affichent les **meilleures performances en ventes**.
- Cela peut être lié à la **maturité des consoles de l’époque** (Wii, PS3, Xbox 360) et à une explosion de la production de jeux.

---

### 🔄 Corrélations observées
- Il existe une **corrélation modérée** entre les **notes** et les **ventes globales** : les jeux bien notés vendent un peu mieux.
- Il y a une **forte corrélation négative** entre l’année de sortie et les ventes, probablement due à l'ancienneté (les vieux jeux ont eu plus de temps pour vendre).
